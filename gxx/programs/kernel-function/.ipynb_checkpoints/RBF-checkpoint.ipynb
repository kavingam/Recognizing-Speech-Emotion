```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the Euclidean distance between two vectors
def euclidean_distance(x, x_prime):
    return np.linalg.norm(x - x_prime)

# Gaussian RBF Kernel function
def gaussian_kernel(x, x_prime, sigma):
    dist = euclidean_distance(x, x_prime)
    return np.exp(-dist ** 2 / (2 * sigma ** 2))

# Example vectors (data points)
x = np.array([1.0, 2.0])  # Point 1 (2D)
x_prime = np.array([3.0, 4.0])  # Point 2 (2D)
sigma = 1.0  # Sigma parameter for Gaussian kernel

# Compute the Gaussian kernel between x and x_prime
kernel_value = gaussian_kernel(x, x_prime, sigma)
print(f"Gaussian RBF Kernel between x and x_prime: {kernel_value}")

# Generate a range of distances to plot the Gaussian kernel values
distances = np.linspace(0, 10, 100)  # Generate distances from 0 to 10
kernel_values = np.exp(-distances ** 2 / (2 * sigma ** 2))

# Plot the Gaussian kernel values as a function of distance
plt.plot(distances, kernel_values, label=f'Sigma = {sigma}')
plt.title("Gaussian RBF Kernel as a Function of Distance")
plt.xlabel("Distance")
plt.ylabel("Kernel Value")
plt.legend()
plt.grid(True)
plt.show()
```
